## Tarea 2: Queries con SQL

### Ejercicio 1

##### Muestra las diferentes plataformas de lanzamiento:

```python
#Rellena la variable query con tu consulta
query = """ SELECT DISTINCT launchSite from SPACEXTBL;"""
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 2

##### Muestra 5 lanzamientos cuya plataforma empiece por 'CCS'


```python
query = """ SELECT LaunchSite, FlightNumber from SPACEXTBL where LaunchSite like '%CCS%' limit 5;"""
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 3

##### Mustra el la masa total transportada por los lanzamientos realizados donde el cliente es NASA (CRS)


```python
query = """ SELECT sum(payloadmass) from SPACEXTBL where customers like '%(NASACRS)%' ;"""
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 4

##### Muestra la masa media tranportada por lanzamientos con serie que comienza por B1

```python
query = """ SELECT serial, avg(Payloadmass) from SPACEXTBL  where serial like '%B1%';"""
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 5

##### Muestra la fecha donde se realizo el primer aterrizaje exitoso sobre terreno (TRUE RTLS)

*Pista:Utiliza la función min*


```python
query = """ SELECT Date, FlightNumber, LandingOutcome from SPACEXTBL where LandingOutcome like '%TRUE RTLS%'  order by date limit 1 ;"""
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 6

##### Muestra las series de los cohetes que aterrizaron sobre portaviones (TRUE ASDS) con una masa entre 4000 y 6000

```python
query = """ SELECT serial ,FlightNumber, LandingOutcome from SPACEXTBL where LandingOutcome like '%TRUE ASDS%' AND Payloadmass BETWEEN 4000 and 6000;"""
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 7

##### Muestra el numero de lanzamientos que completaron su mision correctamente y aquellos que fallaron

```python
query = """ SELECT COUNT(FlightNumber) , Class from SPACEXTBL group by class;"""
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 8

##### Muestra las series de los cohetes que transportaron la masa máxima

```python
query = """ SELECT serial, BoosterVersion, Payloadmass from SPACEXTBL Where Payloadmass >= 15600 group by serial;"""
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 9

##### Muestra los el resultados del aterrizaje, la serie del cohete y las plataforma de lanzamiento de aquellos lanzamientos fallidos o no intentados en el año 2015


```python
query = """ SELECT serial, BoosterVersion, Date, Class, LaunchSite from SPACEXTBL Where Date=2015 and class=0;"""
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 10

##### Muestra los resultados y la frecuencia de estos resultados entre 2010-06-04 y 2017-03-20 ordenados en orden descendiente según la frecuencia


```python
query = """ SELECT serial, BoosterVersion, Date, Class, reusedcount from SPACEXTBL Where Date between '2010-06-04' and '2017-03-20' order by reusedcount desc"""
df  = pd.read_sql(query, con = engine )
df
```
